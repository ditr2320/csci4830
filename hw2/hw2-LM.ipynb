```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
import math
```

```python
def leaky(parameters):
    
    beta, gamma = parameters
    
    dt = 0.1
    # lists to store SIR values & initial conditions
    sList = []
    sList.append(0.49)
    iList = []
    iList.append(0.01)
    rList = []
    rList.append(0)
    vList = []
    vList.append(0.50)
    # list to store time
    time = np.arange(0, 250, dt).tolist()
    
    for i in range(0,len(time)-1):
        s = sList[-1] - ((beta * sList[-1] * iList[-1])) * dt 
        i = iList[-1] + (((beta * sList[-1] * iList[-1]))+ (beta * vList[-1]* iList[-1]*(0.2)) - (gamma * iList[-1])) * dt 
        r = rList[-1] + (gamma * iList[-1]) * dt 
        v = vList[-1] - ((beta * vList[-1] * iList[-1])*(0.2)) * dt 
        
        sList.append(s)
        iList.append(i)
        rList.append(r)
        vList.append(v)
    return((time,sList,iList,rList,vList))
        
```

```python
def plot(time, sList, iList, rList,vList,r0):
    fig, ax = plt.subplots(nrows=1,ncols=1)
    ax.plot(time,sList,label='s',color="blue")
    ax.plot(time,iList,label='i',color="red")
    ax.plot(time,rList,label='r',color="black")
    ax.plot(time,vList,label='v0',color="orange")
    ax.legend(loc='center right')
    ax.set_xlim(left=0)
    ax.set_ylim(bottom=0)
    ax.set_ylim(top=0.8)
    ax.set_xlabel('time')
    ax.set_ylabel('population')
    ax.set_title(r"$r_{0} = $" + str(r0))
```

```python
parameters = (3/14,1/14) # parameters (beta, gamma) if R0 = 3
time, sList, iList, rList,vList  = leaky(parameters)
plot(time,sList,iList,rList,vList,3)
rList[-1]
```
