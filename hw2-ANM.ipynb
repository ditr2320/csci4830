```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
import math
```

```python
def allOrNothing(parameters):
    
    beta, gamma = parameters
    
    dt = 0.1
    # lists to store SIR values & initial conditions
    sList = []
    sList.append(0.49)
    iList = []
    iList.append(0.01)
    rList = []
    rList.append(0)
    v0List = []
    v0List.append(0.10)
    vAList = []
    vAList.append(0.40)

    # list to store time
    time = np.arange(0, 250, dt).tolist()
    
    for i in range(0,len(time)-1):
        s = sList[-1] - ((beta * sList[-1] * iList[-1])) * dt 
        i = iList[-1] + (((beta * sList[-1] * iList[-1]))+ (beta * v0List[-1]* iList[-1]) - (gamma * iList[-1])) * dt 
        r = rList[-1] + (gamma * iList[-1]) * dt 
        v0 = v0List[-1] - ((beta * v0List[-1] * iList[-1])) * dt 
        vA = vAList[-1]
        
        sList.append(s)
        iList.append(i)
        rList.append(r)
        v0List.append(v0)
        vAList.append(vA)
    return((time,sList,iList,rList,v0List,vAList))
        
```

```python
def plot(time, sList, iList, rList, v0List,vAList,r0):
    fig, ax = plt.subplots(nrows=1,ncols=1)
    ax.plot(time,sList,label='s',color="blue")
    ax.plot(time,iList,label='i',color="red")
    ax.plot(time,rList,label='r',color="black")
    ax.plot(time,v0List,label='v0',color="orange")
    ax.plot(time,vAList,label='vA',color="green")
    ax.legend(loc='center right')
    ax.set_xlim(left=0)
    ax.set_ylim(bottom=0)
    ax.set_ylim(top = 0.6)
    ax.set_xlabel('time')
    ax.set_ylabel('population')
    # ax.set_title('All-Or-Nothing Model')
    ax.set_title(r"$r_{0} = $" + str(r0))
```

```python
parameters = (3/14,1/14) # parameters (beta, gamma) if R0 = 3
time, sList, iList, rList,v0List,vAList  = allOrNothing(parameters)
plot(time,sList,iList,rList,v0List,vAList,3)
rList[-1]
```
